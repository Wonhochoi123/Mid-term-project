#### **Task 6**: How many states cover 50% of US air traffic? 

```python
#Task 6: How many states cover 50% of US air traffic?
origin=df.groupby('origin_city_name').size().to_frame()
origin['state']=origin.index.str[-2:]
origin_n=origin.groupby('state').sum().sort_values(0, ascending= False)
origin_n.reset_index(inplace=True)
acc=[]
for i in range(53):
    acc.append(origin_n[:(i+1)].sum()[0])
origin_n['acc']= acc
# we have to add one cuz we want the number of states that COVERS. so the 'acc' should be bigger than 50% of it's entirity
origin_n[(origin_n['acc']<(origin_n[0].sum())/2)].count()[0]+1
```

#### **Task 8**: When (which hour) do most 'LONG', 'SHORT', 'MEDIUM' haul flights take off?

```python
# creating a new column that labels how long the flight is.
task8 = pd.DataFrame(df['crs_elapsed_time'])
task8['crs_dep_time'] = df['crs_dep_time']
task8['type_sml'] = ""
def New(Row):
    if Row['crs_elapsed_time']<180:
        return 'SHORT'
    elif Row['crs_elapsed_time']>=180 and Row['crs_elapsed_time']<360:
        return 'MEDIUM'
    else :
        return 'LONG'
task8['type_sml']=task8.apply(New, axis=1)
```

```python
#getting the hour valued column.
def hr_convert(row):
    return str(row['crs_dep_time'])[:-2]

task8['dep_hr']=task8.apply(hr_convert, axis=1)
task8
```

```python
# those hours are the most common hours to take off for each length of it
task8.groupby('type_sml').dep_hr.agg(pd.Series.mode).to_frame()
```

#### **Task 9**: Find the top 10 the bussiest airports. Does the biggest number of flights mean that the biggest number of passengers went through the particular airport? How much traffic do these 10 airports cover?

```python
# getting total number of landing and take offs then adding those up, getting the total traffic of each airports.
OR_TRAFFIC=df[['origin','flights']].groupby('origin').sum()[['flights']]
OR_TRAFFIC=OR_TRAFFIC.rename(columns = {'flights':'or_fl'})
DE_TRAFFIC=df[['dest','flights']].groupby('dest').sum()[['flights']]
DE_TRAFFIC=DE_TRAFFIC.rename(columns = {'flights':'de_fl'})
TRAFFIC=pd.concat([DE_TRAFFIC, OR_TRAFFIC], axis=1).fillna(0)

TRAFFIC['total_tr']=TRAFFIC['de_fl']+TRAFFIC['or_fl']
TRAFFIC=TRAFFIC.sort_values(by='total_tr',ascending=False)
TRAFFIC[:10]
```

```python
# Top 10 covers this much of total us air traffic
(TRAFFIC[:10].total_tr.sum())/(TRAFFIC.total_tr.sum())*100
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = (list(TRAFFIC.index))
sizes = list(TRAFFIC['total_tr'])
explode = [i*0+0.1 for i in range(10)]+[a*0 for a in range(len(list(TRAFFIC.index))-10)] 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90, textprops={'fontsize': 25})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

#### **Task 10**: Do bigger delays lead to bigger fuel comsumption per passenger?
